```python
import re

def parse_value(x):
    x = str(x).strip().lower()
    
    # Case 1: Range like "10-15" or "9.5-10"
    if re.match(r"^\d+(\.\d+)?\s*-\s*\d+(\.\d+)?$", x):
        a, b = re.split(r"\s*-\s*", x)
        return (float(a) + float(b)) / 2   # midpoint in %
    
    # Case 2: "9 ± 0.5"
    if "±" in x:
        num, err = x.split("±")
        return float(num.strip())  # central value in %
    
    # Case 3: Percentages ("7 - 10%", "12% wb")
    if "%" in x:
        nums = re.findall(r"[\d\.]+", x)
        if len(nums) == 2:   # range like "7 - 10%"
            return (float(nums[0]) + float(nums[1])) / 2
        else:
            return float(nums[0])
    
    # Case 4: Inequality "<14"
    if x.startswith("<"):
        return float(x[1:])  # keep as % value
    
    # Case 5: FTMB values (convert fraction → % wb)
    if "ftmb" in x:
        match = re.search(r"[\d\.]+", x)
        if match:
            return float(match.group(0)) * 100.0  # convert to %
    
    # Default: Extract number if nothing else
    match = re.search(r"[\d\.]+", x)
    if match:
        return float(match.group(0))
    
    return None
```

```python
master_df['Moisture content (% wb)'] = master_df['Moisture content (% wb)'].apply(parse_value)
master_df 
```

```python
import pandas as pd
import re
import numpy as np


def parse_particle_size(x):
    x = str(x).strip().lower()
    
    # 1. Handle "or" case: take average of the two options
    if "or" in x:
        nums = re.findall(r"[\d\.]+", x)
        return np.mean([float(n) for n in nums])
    
    # 2. Handle "x" case: multiply or just average dimensions
    if "x" in x:
        nums = re.findall(r"[\d\.]+", x)
        return np.mean([float(n) for n in nums])  # central value from dimensions
    
    # 3. Handle multiple ranges separated by commas
    if "," in x:
        nums = re.findall(r"[\d\.]+", x)
        return np.mean([float(n) for n in nums])
    
    # 4. Handle ranges "a - b"
    if re.search(r"\d+\s*-\s*\d+", x):
        nums = re.findall(r"[\d\.]+", x)
        if len(nums) >= 2:
            return (float(nums[0]) + float(nums[1])) / 2  # midpoint
    
    # 5. Handle values with parentheses (e.g. 3.9 - 6.8 (6.1))
    if "(" in x and ")" in x:
        nums = re.findall(r"[\d\.]+", x)
        return float(nums[-1])  # use the number in brackets as average
    
    # 6. Default: extract first number
    nums = re.findall(r"[\d\.]+", x)
    if nums:
        return float(nums[0])
    
    return None


```

```python
master_df['Initial avg. particle size, (mm)'] = master_df['Initial avg. particle size, (mm)'].apply(parse_particle_size)
master_df 
```

```python
import pandas as pd
import re

def parse_motor_size(x):
    x = str(x).strip().lower()
    
    # Extract number
    match = re.search(r"[\d\.]+", x)
    if not match:
        return None
    value = float(match.group(0))
    
    # Handle units
    if "hp" in x:
        return value * 0.7457   # convert hp → kW
    elif "(w" in x:  # e.g. (W)
        return value / 1000.0   # convert W → kW
    else:
        return value  # assume already in kW
    
```

```python
master_df['Motor size, kW'] = master_df['Motor size, kW'].apply(parse_motor_size)
master_df 
```

```python
import pandas as pd
import re
import numpy as np

def parse_energy(x):
    x = str(x).strip().lower()
    
    # Case 1: J/g → convert to kJ/kg
    if "j/g" in x:
        num = re.findall(r"[\d\.]+", x)
        if num:
            return float(num[0])  # 1 J/g = 1 kJ/kg
    
    # Case 2: Range "a - b"
    if re.search(r"\d+\s*-\s*\d+", x):
        nums = re.findall(r"[\d\.]+", x)
        if len(nums) >= 2:
            return (float(nums[0]) + float(nums[1])) / 2  # midpoint in kJ/kg
    
    # Default: already in kJ/kg
    nums = re.findall(r"[\d\.]+", x)
    if nums:
        return float(nums[0])
    
    return None

```

```python
master_df['Sp. Grinding energy, kJ/kg'] = master_df['Sp. Grinding energy, kJ/kg'].apply(parse_energy)
master_df 
```

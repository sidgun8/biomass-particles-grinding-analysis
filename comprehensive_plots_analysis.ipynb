```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

```

```python
# Load the cleaned data
df = pd.read_csv("final_cleaned_data.csv")

# Capitalize all string values
df = df.map(lambda x: x.upper() if isinstance(x, str) else x)

# Create directories for saving graphs
os.makedirs("graphs", exist_ok=True)
os.makedirs("graphs_no_outliers", exist_ok=True)

df.head()

```

# Final Particle Size (Grouped by Mill Type)



## 1. Final Particle Size vs Moisture Content (Grouped by Mill Type)



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data cleaning
plot_df = df.dropna(subset=["Final particle size (mm)", "Moisture content (% wb)", "Grinder type"]).copy()

# Standardize text
plot_df["Grinder type"] = plot_df["Grinder type"].str.strip().str.title()

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=plot_df,
    x="Moisture content (% wb)",
    y="Final particle size (mm)",
    hue="Grinder type",
    s=80,
    alpha=0.85
)

plt.title("Final Particle Size vs Moisture Content by Grinder Type", fontsize=14)
plt.xlabel("Moisture Content (% wb)", fontsize=12)
plt.ylabel("Final Particle Size (mm)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(title="Grinder Type", fontsize=9)
plt.tight_layout()
plt.savefig("graphs/final_particle_size_vs_moisture_by_mill_type.png", dpi=300, bbox_inches='tight')
plt.show()

```

```python
# Outlier removal function using IQR method
def remove_outliers_iqr(data, col):
    """Remove outliers from a column using the IQR method."""
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[col] >= lower) & (data[col] <= upper)]

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data cleaning
plot_df = df.dropna(subset=["Moisture content (% wb)", "Sp. Grinding energy, kJ/kg", "Biomass"]).copy()

# Standardize text
plot_df["Biomass"] = plot_df["Biomass"].str.strip().str.title()

# Remove outliers
plot_df = remove_outliers_iqr(plot_df, "Moisture content (% wb)")
plot_df = remove_outliers_iqr(plot_df, "Sp. Grinding energy, kJ/kg")

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=plot_df,
    x="Moisture content (% wb)",
    y="Sp. Grinding energy, kJ/kg",
    hue="Biomass",
    s=80,
    alpha=0.85
)

plt.title("Moisture Content vs Specific Grinding Energy by Biomass Type (Outliers Removed)", fontsize=14)
plt.xlabel("Moisture Content (% wb)", fontsize=12)
plt.ylabel("Specific Grinding Energy (kJ/kg)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(title="Biomass Type", fontsize=9)
plt.tight_layout()
plt.savefig("graphs_no_outliers/specific_grinding_energy_vs_moisture_by_biomass_type.png", dpi=300, bbox_inches='tight')
plt.show()

```
